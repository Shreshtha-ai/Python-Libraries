```python
import pandas as pd
print(pd.__version__)
```

```python
data = [100.1, 120.2 , 200.4]
series = pd.Series(data , index = ["A","B","C"])
series.loc["C"] = 246 # LOC is location by label 
print(series)
print(series.loc["A"])

print(series.iloc[0]) #iloc is used for location by integer 


```

```python
data = [100.1, 120.2 , 200.4 , 300 , 450]

series = pd.Series(data , index = ["A","B","C","D","E"])
print(series[series>= 200])

#Series: One-dimensional ndarray with axis labels
 
```

```python
calories = {"Day1":1750, "Day2":1800, "Day3":2000, "Day4":2100, "Day5":1760}
series = pd.Series(calories)
print(series[series<2000])
```
