```python
import pandas as pd

data = {
    "Name": ["Spongebob","Patrick","Squidward"],
    "Age": [30,45,50]

}
df = pd.DataFrame(data, index = ["Employee1","Employee2","Employee3"])
print(df)
print(df.loc["Employee1"])
print(df.iloc[2])

#Creating new column
df["Job"] = ["Cook","N/A","Cashier"]
print(df)

#Creating new row
new_row = pd.DataFrame([{"Name":"Sandy","Age":43,"Job":"Enginner"}],
                        index = ["Employee4"])

df = pd.concat([df , new_row])
print(df)
```
